```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
data = pd.read_csv('averagewaitingtime.csv')

# Extract day from the date
data['day'] = pd.to_datetime(data['date']).dt.day

# Group by the extracted day to ensure we are working with unique days
# In case of multiple entries per day, we'll average the waiting time.
daily_average = data.groupby('day')['request_to_pickup'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(daily_average['day'], daily_average['request_to_pickup'], color='violet')
plt.xlabel('Days of Jan month')
plt.ylabel('Average Waiting Time')
plt.title('Average Waiting Time by Day in January')
plt.xticks(daily_average['day'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
