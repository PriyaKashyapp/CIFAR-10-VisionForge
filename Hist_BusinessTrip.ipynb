```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Creating the DataFrame from the provided data
# data = {
#     'business': ['Lyft-4', 'Uber-5', 'Uber-4', 'Lyft-3', 'Uber-2', 'Lyft-1', 'Uber-3', 'Uber-1', 'Lyft-2', 'Lyft-5'],
#     'count': [8173, 14276372, 13995860, 8444, 13280761, 6887, 14554308, 13579077, 6491, 9491]
# }
# trip_df = pd.DataFrame(data)

trip_df = pd.read_csv('businessMtrips.csv')


#logarithmic transformation to the 'count' column
trip_df['log_count'] = np.log10(trip_df['count'])

plt.figure(figsize=(10, 6))
plt.bar(trip_df['business'], trip_df['count'], color='skyblue')
plt.xlabel('Business per month')
plt.ylabel('Number of Trips')
plt.title('Count the number of trips for each business in each month')
plt.xticks(rotation=90)
plt.tight_layout()


y_ticks = plt.gca().get_yticks()
plt.gca().set_yticklabels([f"{10**val:.0f}" for val in y_ticks])

plt.show()
```

**Calculate the monthly profit for each business**

**Explaination**
platformprofit.csv name of csv file used for this task where I created a bar chart that visualizes the earnings of drivers for various businesses across different months. Positive earnings are shown with sky-blue bars above the x-axis, while negative earnings are depicted with red bars extending below the x-axis for business Uber and Lyft. The y-axis uses a symmetric logarithmic scale to handle **both large positive and negative values effectively**, allowing for a clear representation of data.

```python
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv('platformprofit.csv')

# Define colors for positive (blue) and negative (red) values
colors = ['red' if x < 0 else 'skyblue' for x in df['profit_sum']]

# Plotting the bar chart
plt.figure(figsize=(14, 8))
bars = plt.bar(df['business'], df['rideshare_profit'], color=colors)

# Setting the y-axis to a symlog scale to accommodate large and small values
plt.yscale('symlog')

# Labeling the axes and title
plt.xlabel('Business/Month')
plt.ylabel("Driver's Earnings")
plt.title("Calculate the monthly profit for each business")

# Rotating x-tick labels for better readability
plt.xticks(rotation=90)

# Formatting the y-axis to display numbers with commas for readability
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: f"${int(x):,}"))

# Adding grid lines for better visualization of data points
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Tight layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```

**Calculate the driver's earnings for each business in each month**

**Explanation**
driverearning.csv is the csv file used to depict the driver's earning per month for each business (uber/lyft), chart for Lyft in light blue and another for Uber in light pink, labeling y axis as driver's earning.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'data_updated' DataFrame contains the updated data provided
data_updated = pd.DataFrame({
    'business_concat': ['Lyft-4', 'Uber-5', 'Uber-4', 'Lyft-3', 'Uber-2', 'Lyft-1', 'Uber-3', 'Uber-1', 'Lyft-2', 'Lyft-5'],
    'driver_pay_sum': [297815.3795, 3.13E+08, 2.95E+08, 310276.5494, 2.52E+08, 239932.2594, 2.96E+08, 2.50E+08, 234875.5296, 360408.0897]
})

# Separating the data into two DataFrames, one for Lyft and one for Uber
lyft_data = data_updated[data_updated['business'].str.contains('Lyft')]
uber_data = data_updated[data_updated['business'].str.contains('Uber')]

# Plotting Lyft
plt.figure(figsize=(10, 6))
plt.bar(lyft_data['business'], lyft_data['driver_total_pay'], color='lightblue')
plt.xticks(rotation=45, ha="right")
plt.xlabel('Month')
plt.ylabel("Driver's Earnings")
plt.title("Lyft Driver's Earnings")
plt.tight_layout()

# Showing the Lyft plot
plt.show()

# Plotting Uber
plt.figure(figsize=(10, 6))
plt.bar(uber_data['business'], uber_data['driver_total_pay'], color='lightpink')
plt.xticks(rotation=45, ha="right")
plt.xlabel('Month')
plt.ylabel("Driver's Earnings")
plt.title("Uber Driver's Earnings")
plt.tight_layout()

# Showing the Uber plot
plt.show()
```
